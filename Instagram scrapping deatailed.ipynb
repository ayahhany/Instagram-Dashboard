```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait
import time
```

```python
#specify the path to chromedriver.exe
driver = webdriver.Chrome()

#open the webpage
driver.get("http://www.instagram.com")
```

```python
for user in usernames[:15]:
    print(user)
    searchbox = driver.find_element(By.XPATH, "//input[@placeholder='Search']")
    searchbox.clear()
    keyword = user
    searchbox.send_keys(keyword)
    time.sleep(1) 
    element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located(
            (By.XPATH, '//*[@id="react-root"]/section/nav/div[2]/div/div/div[2]/div[3]/div/div[2]/div/div[1]/a'))
    )
    element.click()
    time.sleep(2)
    driver.refresh()
    time.sleep(3)
    html = driver.find_element_by_tag_name('html')
    for i in range(60):
        html.send_keys(Keys.PAGE_DOWN)
    # get 3 photos at a time
    '''for i in range(1,4):
        post = driver.find_element(By.XPATH, '')'''
    try: 
        links = driver.find_elements_by_tag_name('a')
        links = [link.get_attribute('href') for link in links]
    except:
        pass
    posts_links = []
    for link in links:
        try:
            if 'https://www.instagram.com/p' in link:
                posts_links.append(link)
        except :
            
            print("link was not found")
            continue
    print(len(posts_links))
    posts_links_all.extend(posts_links[:20])
    time.sleep(5)
    print("Done")
    print('-------------')

    
```

```python
likes = []
comments = []
captions = []
datetimes = []
post_types = []
```

```python
posts_links_all[0]
```

```python
for link in posts_links_all:
    #webbrowser.open(link)
    driver.get(link)
    time.sleep(2)
    driver.refresh()
    time.sleep(3)
    try:
        if driver.find_element(By.XPATH, '//*[@id="react-root"]/section/main/div/div[1]/article/div/div[2]/div/div[2]/section[2]/div/div[2]/div/a/div/span'):
            likes_number = driver.find_element(By.XPATH, '//*[@id="react-root"]/section/main/div/div[1]/article/div/div[2]/div/div[2]/section[2]/div/div[2]/div/a/div/span').text
            likes_number = int(likes_number.replace(',', '')) +1
    except:
            likes_number = 'unknown'
    likes.append(likes_number)
    
    try:
        if driver.find_element(By.XPATH, '//*[@id="react-root"]/section/main/div/div[1]/article/div/div[2]/div/div[2]/div[1]/ul/div/li/div/div/div[2]/div[1]/span'):
            caption = driver.find_element(By.XPATH, '//*[@id="react-root"]/section/main/div/div[1]/article/div/div[2]/div/div[2]/div[1]/ul/div/li/div/div/div[2]/div[1]/span').text
    except:
            caption = ''

    captions.append(caption)
    datetime = driver.find_element(By.XPATH, '//*[@id="react-root"]/section/main/div/div[1]/article/div/div[2]/div/div[2]/div[2]/div/a/div/time').get_attribute('datetime')
    datetimes.append(datetime)
    counter = 0
    try:
        if (driver.find_element(By.XPATH,'//*[@id="react-root"]/section/main/div/div[1]/article/div/div[1]/div/div/div[1]/div/div/video')):
            post_type = 'clip'
    except:
        counter+=1
    try:
        if (driver.find_element(By.XPATH, '//*[@id="react-root"]/section/main/div/div[1]/article/div/div[1]/div/div[1]/div[2]/div/button/div')):
            post_type = 'carousel'
    except:
        counter+=1
    
    if counter > 1:
        post_type = 'photo'
    post_types.append(post_type)
    
    

    
```

```python
detailed["likes"] = likes
```

```python
detailed["type"] = post_types
```

```python
detailed["datetimes"] = datetimes
```

```python
detailed["captions"] = captions
```
